卡方分配(χ²-distribution)  
卡方分布是一種特殊的伽瑪分布  
**機率密度函數**  
${\displaystyle f_{k}(x)={\frac {1}{2^{\frac {k}{2}}\Gamma ({\frac {k}{2}})}}x^{{\frac {k}{2}}-1}e^{\frac {-x}{2}}}$  
其中${\displaystyle x\geq 0}{\displaystyle x\geq 0}$，當${\displaystyle x\leq 0}$${\displaystyle x\leq 0}$時${\displaystyle f_{k}(x)=0}f_{k}(x)=0$。這裡Γ代表Gamma函數。

```python
import numpy as np
from scipy.stats import chi2
import matplotlib.pyplot as plt
fontparams = {'font.size': 13, 'font.weight':'bold', 'font.family':'arial', 'font.style':'italic'}
plt.rcParams.update(fontparams)
labelparams = {'size': 14, 'weight':'semibold', 'family':'serif', 'style':'italic'}
fig, axes = plt.subplots(2, 1,figsize=(8,7), constrained_layout=True)
ax1 = axes[0]
ax2 = axes[1]

xlim = [0, 15]
x = np.linspace(xlim[0], xlim[1], 1000)
df = np.arange(1, 10, 2)
ax1.set_xlim([xlim[0], xlim[1]] )
ax1.set_ylim([0, 0.4])

for i in df:
    label = "k = {}".format(i)
    y=chi2.pdf(x, i)
    ax1.plot(x, y, lw=1, label = label ,linewidth=4 ,alpha = 0.6)
ax1.set_xlabel('x',labelparams)
ax1.set_ylabel('f(x)',labelparams)
ax1.set_yticks([0, 0.1, 0.2, 0.3 ,0.4])
ax1.legend()
#############################
xlim = [0, 50]
x = np.linspace(xlim[0], xlim[1], 1000)
df = np.arange(4, 32, 2)

for i in df:
    y=chi2.pdf(x, i)
    ax2.plot(x,y, lw=2.35, color='#005A8D', )

ax2.set_xlabel('x',labelparams)
ax2.set_ylabel('f(x)',labelparams)
ax2.set_xlim([xlim[0], xlim[1]])
ax2.set_ylim([0, 0.2])
ax2.set_yticks([0, 0.1, 0.2])
ax2.text(35, 0.1, 'Degrees of Freedom\n\n For k = 4 to 32, Step 2', fontsize=16, color='black', ha='center')
plt.suptitle('Chi-Square Distributions \n for Different Degrees of Freedom', weight = 'bold' ,ha="center",y=1.045 )


fig.tight_layout()
plt.savefig('chi.eps', format='eps',bbox_inches='tight', dpi=200)
from google.colab import files
files.download('chi.eps')
plt.show()
```
